```python
path_to_folder = userdata.get('path_to_folder_ppsh') + '/Dataset/Parsing/'
```

```python
def read(path):
  with open(path, 'r') as file:
    return file.read()
```

```python
def form_data(path, reg, topic):
    text = read(path_to_folder + path)
    pattern = re.compile(reg, re.DOTALL)
    task = pattern.findall(text)
    data = pd.DataFrame({"Задание": task,
                        "Тема": topic})
    return data
```

# Геометрия

```python
data = form_data('Геометрия/2016_04_28_gomotetia.txt',
          r'\d+\..*?(?=\d+\..*?|$)',
          "Геометрия")[2:]
```

```python
data = pd.concat([data, form_data('/Геометрия/геома.txt',
                                  r'Задача \d+\. \([^)]*\).*?(?=Задача|$)',
                                  'Геометрия')])
```

# Комбинаторика

```python
text = read(path_to_folder + 'Комбинаторика/apc.txt')

pattern = re.compile(r'(?<!20\b)\b\d{1,2}[.)]\s*(?:\([^)]*\))?.*?(?=(?:\b\d+[.)])|\b\d+\s*-\s*\d+[.)]|\bРешение|20|66.\b|\d+\.)', re.DOTALL)

task = pattern.findall(text)
data = pd.concat([data, pd.DataFrame({"Задание": task,
                     "Тема": "Комбинаторика"})])


text = read(path_to_folder + 'Комбинаторика/sumprod.txt')

pattern = re.compile(r'(?<!20\b)\b\d{1,2}[.)]\s*(?:\([^)]*\))?.*?(?=(?:\b\d+[.)])|\b\d+\s*-\s*\d+[.)]|\bРешение|20|66.\b|\d+\.)', re.DOTALL)

task = pattern.findall(text)
data = pd.concat([data, pd.DataFrame({"Задание": task,
                     "Тема": "Комбинаторика"})])
```
