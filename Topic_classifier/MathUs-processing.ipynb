```python
path_to_folder = userdata.get('path_to_folder_ppsh') + '/Dataset/Parsing/'
```

```python
def read(path):
  with open(path, 'r') as file:
    return file.read()
```

```python
def form_data(path, reg, topic):
    text = read(path_to_folder + path)
    pattern = re.compile(reg, re.DOTALL)
    task = pattern.findall(text)
    data = pd.DataFrame({"–ó–∞–¥–∞–Ω–∏–µ": task,
                        "–¢–µ–º–∞": topic})
    return data
```
